## Functions of a random variable

Given a random variable $x$ with probability density function $p_x(x)$ and a function $f(x)$, consider the quantity:
$$y = f(x)$$
Clearly $y$ is also a random variable with some probability density function $p_y(y)$. The relationship between $x$ and $y$ can be visualized as follows:
1. Draw a collection of $n$ samples $X = \{x_1, ...x_n\}$ from the distribution described by $p_x$.
2. Compute the corresponding vector of $y$ values $Y =  \{f(x_1), ... f(x_n)  \}$ and make a scatterplot of the $n$ points $\{(x_1, y_1),...(x_n,y_n)\}$ in the $x$-$y$ plane. 
3. Below the $x$-axis of the scatterplot, draw a histogram of the variables $x_1 ... x_n$, where the bins align with the scatterplot's x axis.
4. To the right of the scatterplot, draw a histogram of the variables $y_1 ... y_n$ which is rotated 90 degrees clockwise so that the bins align with the scatterplot's y-axis.

Below I've written a function `sample_cdf` which takes as input a function `cdf` and an integer `n` and outputs an array of `n` realizations of a random variable given by the distribution `cdf`. I've also written a function `double_hist` which takes two arrays of the same length (`X` and `Y`) and makes the double-histogram-scatterplot described above.

```python
def sample_cdf(cdf, n, domain=(-10,10)):
    X = np.zeros(n)
    x_range = np.linspace(domain[0],domain[1],1000)
    cdf_vals = cdf(x_range)
    x_unf = np.random.rand(n)
    for i in range(n):
        X[i] =  x_range[abs(np.diff(np.sign(cdf_vals - x_unf[i])))/2]
    return X

def double_hist(x_values, y_values):
    global fig, ax_f, ax_x, ax_y
    fig = plt.figure()
    ax_f = plt.subplot2grid((4, 4), (0, 0), colspan=3, rowspan=3)
    ax_x = plt.subplot2grid((4, 4), (3, 0), colspan=3)
    ax_y = plt.subplot2grid((4, 4), (0, 3), rowspan=3)
    ax_x.hist(x_values, bins=100, density = True)
    ax_f.scatter(x_values, y_values)
    ax_y.hist(y_values, bins=100, orientation='horizontal',
              density = True)
    ax_y.set_autoscale_on(False)
    ax_x.set_autoscale_on(False)
    plt.tight_layout()
```

Let's try it out for the a normally distributed $X$ and the function $f(x) = x^2$.

```python
def f(x):
    return (x+10)**3.

gaussian_pdf = stats.norm(loc=0, scale=2).pdf
X = stats.norm(loc=0, scale=2).rvs(size=1000)
double_hist(X, f(X))

# add a few extra lines
ax_x.vlines([-1,1],0,1)
ax_y.hlines([f(-1),f(1)],0,1)
ax_f.vlines([-1,1],[0,0],[f(-1),f(1)])
ax_f.hlines([f(-1),f(1)],[-1,1],max(X))
ax_f.plot([-1,1],[f(-1),f(1)], color='red', linewidth=2)

plt.show()
```

In the plot above I've added a few extra lines. The lines on the $x$-histogram (bottom) mark the edges of the interval $[-1,1]$ and the lines on the $y$-histogram (right) mark the edges of the interval $[f(-1), f(1)]$. Both intervals are marked on the graph of $f(x)$. It is not hard to see that the number of $y$-values in the interval $[f(a), f(b)]$ must equal the number of $x$-values in the interval $[a,b]$. If we assume that the intervals $[a,b]$ and $[f(a),f(b)]$ are small enough that $p_x$ and $p_y$ are approximately constant in the intervals then we can write the following:
$$(b-a)p_x(a) = |f(b) - f(a)|p_y(f(a))$$
And so in the limit we have:
\begin{equation}
\label{inv_rule}
p_y(y) = \frac{p_x[f^{-1}(y)]}{|f'[f^{-1}(y)]|}
\end{equation}
Let's test out this formula by producing the same plot above, but with our prediction for the pdf of $x$ and $y$ plotted over the histograms.

```python
def f(x):
    return (x+10)**3
def f_prime(x):
    return 3*(x+10)**2
def f_inv(y):
    return y**(1./3.) - 10

p_x = gaussian_pdf
p_y = lambda y: p_x(f_inv(y))/abs(f_prime(f_inv(y)))
X = X = stats.norm(loc=0, scale=2).rvs(size=1000)
Y = f(X)
double_hist(X,Y)
ax_x.plot(np.sort(X), p_x(np.sort(X)), color='red')
ax_y.plot(p_y(np.sort(Y)), np.sort(Y), color='red')
plt.show()
```

It works pretty well! In general Equation (\ref{inv_rule}) will not work however, because $f$ may map multiple $x$ values to the same $y$ value, or because its derivative is zero somewhere. We must use Equation (\ref{n_to_1}) instead:

\begin{equation}
\label{n_to_1}
p_y(y) =  \sum_{x: f(x) = y} \frac{p_x(x)}{|f'(x)|}
\end{equation}

```python
def f(x):
    return (x)**2
def f_prime(x):
    return 2*x
def f_pre(y):
    if y==0:
        return 0
    else: 
        return np.array([np.sqrt(y), -np.sqrt(y)])

def p_x(x):
    p1 = stats.norm(loc=-2, scale=1/2).pdf(x)*.75
    p2 = stats.norm(loc=2, scale=1/2).pdf(x)*.25
    return p1+p2

def p_y(y):
    x = f_pre(y)
    return np.sum(p_x(x)/np.abs(f_prime(x)))


X = np.sort(np.append(stats.norm(loc=-2, scale=1/2).rvs(size=750), 
     stats.norm(loc=2, scale=1/2).rvs(size=250)))
Y = f(X)
double_hist(X,Y)
x_plt = np.linspace(min(X),max(X),1000)
y_plt = np.linspace(min(Y),max(Y),1000)
ax_x.plot(X, p_x(X), color='red')
ax_y.plot([p_y(yi) for yi in Y], Y, color='red')
plt.show()
```

Now we consider the inverse problem. That is, given the distribution $p_y$, what can we infer about $p_x$? In the case that $f$ is a one-to-one function, Equation (\ref{inv_rule}) implies the following:

\begin{equation}
\label{inv_rule_2}
p_x(x) = f'(x) p_y[f(x)]
\end{equation}

If $f$ is not one-to-one then it is not possible to guarantee perfect reconstruction of $p_x$ from $p_y$, due to the simple fact that Equation (\ref{n_to_1}) is not invertible. The best we can do is to assume that the summands are equal.

\begin{equation}
\label{1_to_n}
p_x(x) = \frac{f'(x) p_y[f(x)]}{\text{size}\{x': f(x')=f(x)\}}
\end{equation}

## Function Spaces
$L^2[0,1]$ denotes the set of functions which are defined only on the interval $[0,1]$ and which have a convergent Fourier series. It is an example of a vector space of countably infinite dimension. Suppose $O$ is an operator whose input is any function $f$ in $L^2[0,1]$ and whose output $O(f)$ is a real number.

Let $\hat{O}$ denote the best linear approximation to $O$ at some function $\hat{f}$. That is, $\hat{O}(\hat{f}) = O(\hat{f})$
