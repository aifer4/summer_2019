```python
from __future__ import division #(for python 2.7 compatibility)
import par
import modes
import spectra
import numpy as np
import numba
import matplotlib.pyplot as plt
import pickle
```

```python
"""Set the A (scale factor) and K (wavenumber) arrays. 
They can be of any length"""
# It is important to remember that the output array is half the size
# of the input array because RK4 is used to solve the ODE.
A = np.linspace(par.a0, par.a_rec, par.NT)
K = np.linspace(1,1000,par.NK)

wC = np.zeros(par.NT)
cs2C = np.zeros((par.NT,par.NK))
wN = np.ones(par.NT)*1/3
cs2N = np.ones((par.NT,par.NK))*1/3
```

```python
Cl = spectra.get_Cl_3gdm(A, K, wC, cs2C, wN, cs2N)
Cl_err = spectra.get_Cl_err()
ell = spectra.l_list

plt.loglog(ell, Cl)
plt.loglog(ell, Cl_err)
plt.show()
```

```python
theta_cs2C = np.zeros(par.NC**2)

@numba.njit
def square_basis(theta):
    height = par.NT//par.NC
    width = par.NK//par.NC
    out = np.zeros((par.NK, par.NT))
    for i in range(par.NC**2):
        r,c = i//par.NC, i%par.NC
        out[r*width:(r+1)*width, c*height:(c+1)*height] = theta[i]
    out[:,A < par.a0_pca]=0
    return out.T

#@numba.njit
def get_square(n, val):
    height = par.NT//par.NC
    width = par.NK//par.NC
    out = np.zeros((par.NK, par.NT))
    r,c = n//par.NC, n%par.NC
    out[r*width:(r+1)*width, c*height:(c+1)*height] = val
    out[:,A < par.a0_pca]=0
    return out.T

theta_cs2C[31] = .1
plt.imshow(cs2C + get_square(1, 0.05),aspect='auto')
plt.colorbar()
```

```python
ss = 0.01 #(derivative step size)
J = np.zeros((spectra.L, par.NC**2))
for n in range(par.NC**2):
    delta_cs2 = get_square(n, ss)
    Cl_fwd = spectra.get_Cl_3gdm(A, K, wC, cs2C, wN, cs2N+delta_cs2)
    Cl_bwd = spectra.get_Cl_3gdm(A, K, wC, cs2C, wN, cs2N-delta_cs2)
    J[:, n] = (Cl_fwd - Cl_bwd)/(2.*ss)
    if(n % 10 == 0):
        print(str(n) + ' of ' + str(par.NC**2))
```

```python
pickle.dump(J, open("data/J_cs2N_"+str(par.NC)+".p", "wb"))
```

```python
def get_PC(NC):
    Cl_err = spectra.get_Cl_err()
    J = pickle.load(open("data/J_cs2_"+str(NC)+".p", "rb"))
    Sigma = np.diag(Cl_err)


    # just using Identity as covariance matrix for now
    F = np.transpose(J)@Sigma@J
    u, s, v = np.linalg.svd(F)
    D, V = np.linalg.eig(F)
    # principal vectors are currently defined as the rank-ordered left singular vectors
    PV = np.real(u) 
    #PC = B@PV
    plt.show()
    PC = np.zeros((10, par.NK, par.NT))
    for i in range(10):
        PC[i,:,:] = square_basis(NC, PV[:,i])
        #PC[i,:,:] = PC[i,:,:]/np.sum(PC[i,:,:]**2)
    return PC

#J = pickle.load(open("data/J_cs2_"+str(par.NC)+".p", "rb"))
Sigma = np.diag(Cl_err)


# just using Identity as covariance matrix for now
F = np.transpose(J)@Sigma**-1@J
u, s, v = np.linalg.svd(F)
D, V = np.linalg.eig(F)
# principal vectors are currently defined as the rank-ordered left singular vectors
PV = np.real(u) 
#PC = B@PV
plt.show()
PC = np.zeros((10, par.NK, par.NT))
print(np.shape(PV[:,1]))
```
