```python
import params as par
import numpy as np
import matplotlib.pyplot as plt
import pickle
plt.rcParams.update(par.plot_params)
```

```python
τ_solve = pickle.load(open( "data/tau_solve.p", "rb" ) )
τ_clip = par.τ0*2
T = τ_solve[τ_solve>τ_clip]
print(len(T))

n = 0
# sets derivative step size. right now there is data for t=3,4,6,7.
t = 7
Nmax_cos = Nmax_sinc=Nmax_square=Nmax_gaussian=0

plt.title('PC '+str(n+1)+'. Cosine Basis. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
leg = []
for log2NC in range(4,10):
    try:
        PC = pickle.load(open( "data/PC_t-"+str(t)+"_cosine-"+str(2**log2NC)+"_c.p", "rb" ) )
        PC = PC[τ_solve>τ_clip]
        plt.plot(T, PC[:,n])
        leg = leg+[str(2**log2NC) + ' components']
        Nmax_cos = 2**log2NC
    except: 
        pass
plt.legend(leg)
plt.show()

plt.title('PC '+str(n+1)+'. Square Basis. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
leg = []
for log2NC in range(4,10):
    try:
        PC = pickle.load(open( "data/PC_t-"+str(t)+"_square-"+str(2**log2NC)+"_c.p", "rb" ) )
        PC = PC[τ_solve>τ_clip]
        plt.plot(T, PC[:,n])
        leg = leg+[str(2**log2NC) + ' components']
        Nmax_square = 2**log2NC
    except: 
        pass
plt.legend(leg)
plt.show()

plt.title('PC '+str(n+1)+'. Gaussian Basis. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
leg = []
for log2NC in range(4,10):
    try:
        PC = pickle.load(open( "data/PC_t-"+str(t)+"_gaussian-"+str(2**log2NC)+"_c.p", "rb" ) )
        PC = PC[τ_solve>τ_clip]
        plt.plot(T, PC[:,n])
        leg = leg+[str(2**log2NC) + ' components']
        Nmax_gaussian = 2**log2NC
    except: 
        pass
plt.legend(leg)
plt.show()

plt.title('PC '+str(n+1)+'. Sinc Basis. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
leg = []
for log2NC in range(4,10):
    try:
        PC = pickle.load(open( "data/PC_t-"+str(t)+"_sinc-"+str(2**log2NC)+"_c.p", "rb" ) )
        PC = PC[τ_solve>τ_clip]
        plt.plot(T, PC[:,n])
        leg = leg+[str(2**log2NC) + ' components']
        Nmax_sinc = 2**log2NC
    except: 
        pass
plt.legend(leg)
plt.show()



plt.title(r'PC '+str(n+1)+'. All Bases. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
PC_sinc = pickle.load(open( "data/PC_t-"+str(t)+"_sinc-"+str(Nmax_sinc)+"_c.p", "rb" ) )
PC_cosine = pickle.load(open( "data/PC_t-"+str(t)+"_cosine-"+str(Nmax_cos)+"_c.p", "rb" ) )
PC_square = pickle.load(open( "data/PC_t-"+str(t)+"_square-"+str(Nmax_square)+"_c.p", "rb" ) )
PC_gaussian = pickle.load(open( "data/PC_t-"+str(t)+"_gaussian-"+str(Nmax_gaussian)+"_c.p", "rb" ) )
(PC_sinc, PC_cosine, PC_square, PC_gaussian)=\
(PC_sinc[τ_solve>τ_clip,:], PC_cosine[τ_solve>τ_clip], PC_square[τ_solve>τ_clip], PC_gaussian[τ_solve>τ_clip])

plt.plot(T,PC_sinc[:,n])
plt.plot(T,PC_cosine[:,n])
plt.plot(T,PC_square[:,n])
plt.plot(T,PC_gaussian[:,n])
plt.legend(['sinc (%d components)'%Nmax_sinc,'cosine (%d components)'%Nmax_cos,
            'square (%d components)'%Nmax_square,'gaussian (%d components)'%Nmax_gaussian])
plt.show()

plt.title(r'PC '+str(n+1)+'. All Bases. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
print(len(PC_sinc[:,0]))
plt.plot(T,PC_sinc[:,n])
plt.plot(T,PC_cosine[:,n])
plt.plot(T,PC_square[:,n])
plt.plot(T,PC_gaussian[:,n])
plt.legend(['sinc (%d components)'%Nmax_sinc,'cosine (%d components)'%Nmax_cos,
            'square (%d components)'%Nmax_square,'gaussian (%d components)'%Nmax_gaussian])
plt.show()
```
