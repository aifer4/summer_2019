```python
import params as par
from params import *
import numpy as np
import matplotlib.pyplot as plt
import pickle
plt.rcParams.update(par.plot_params)
```

```python
_solve = pickle.load(open( "data/tau_solve.p", "rb" ) )
T = _solve[_solve>0_pca]
print(len(T))
n = 0
# sets derivative step size. right now there is data for t=3,4,6,7.
t = 7
Nmax_cos = Nmax_sinc=Nmax_square=Nmax_gaussian=0

plt.title('PC '+str(n+1)+'. Cosine Basis. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
leg = []
for log2NC in range(4,10):
    try:
        PC = pickle.load(open( "data/PC_t-"+str(t)+"_cosine-"+str(2**log2NC)+".p", "rb" ) )
        plt.plot(_solve, PC[:,n])
        leg = leg+[str(2**log2NC) + ' components']
        Nmax_cos = 2**log2NC
    except: 
        pass
plt.axvline(0_pca,color='black',linestyle='--')
leg = leg + ['basis start']
plt.legend(leg)
plt.savefig('plots/CV_only/' + 'PC-'+str(n+1)+'_cosine.pdf',format='pdf')
plt.show()

plt.title('PC '+str(n+1)+'. Square Basis. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
leg = []
for log2NC in range(4,10):
    try:
        PC = pickle.load(open( "data/PC_t-"+str(t)+"_square-"+str(2**log2NC)+".p", "rb" ) )
        plt.plot(_solve, PC[:,n])
        leg = leg+[str(2**log2NC) + ' components']
        Nmax_square = 2**log2NC
    except: 
        pass
plt.axvline(0_pca,color='black',linestyle='--')
leg = leg + ['basis start']
plt.legend(leg)
plt.savefig('plots/CV_only/' + 'PC-'+str(n+1)+'_square.pdf',format='pdf')
plt.show()

plt.title('PC '+str(n+1)+'. Gaussian Basis. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
leg = []
for log2NC in range(4,10):
    try:
        PC = pickle.load(open( "data/PC_t-"+str(t)+"_gaussian-"+str(2**log2NC)+".p", "rb" ) )
        plt.plot(_solve, PC[:,n])
        leg = leg+[str(2**log2NC) + ' components']
        Nmax_gaussian = 2**log2NC
    except: 
        pass
plt.axvline(0_pca,color='black',linestyle='--')
leg = leg + ['basis start']
plt.legend(leg)
plt.savefig('plots/CV_only/' + 'PC-'+str(n+1)+'_gaussian.pdf',format='pdf')
plt.show()


plt.title(r'PC '+str(n+1)+'. All Bases. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
plt.semilogx()
PC_cosine = pickle.load(open( "data/PC_t-"+str(t)+"_cosine-"+str(Nmax_cos)+".p", "rb" ) )
PC_square = pickle.load(open( "data/PC_t-"+str(t)+"_square-"+str(Nmax_square)+".p", "rb" ) )
PC_gaussian = pickle.load(open( "data/PC_t-"+str(t)+"_gaussian-"+str(Nmax_gaussian)+".p", "rb" ) )
plt.plot(_solve,PC_cosine[:,n])
plt.plot(_solve,PC_square[:,n])
plt.plot(_solve,-PC_gaussian[:,n])
plt.axvline(0_pca,color='black',linestyle='--')
plt.legend(['cosine (%d components)'%Nmax_cos,
            'square (%d components)'%Nmax_square,
            'gaussian (%d components)'%Nmax_gaussian,
           'basis start'])
plt.savefig('plots/CV_only/' + 'PC-'+str(n+1)+'_all_log.pdf',format='pdf')
plt.show()

plt.title(r'PC '+str(n+1)+'. All Bases. Step Size = '+str(10**(1-t)))
plt.xlabel(r'$H_0 \tau$')
PC_cosine = pickle.load(open( "data/PC_t-"+str(t)+"_cosine-"+str(Nmax_cos)+".p", "rb" ) )
PC_square = pickle.load(open( "data/PC_t-"+str(t)+"_square-"+str(Nmax_square)+".p", "rb" ) )
PC_gaussian = pickle.load(open( "data/PC_t-"+str(t)+"_gaussian-"+str(Nmax_gaussian)+".p", "rb" ) )
plt.plot(_solve,PC_cosine[:,n])
plt.plot(_solve,PC_square[:,n])
plt.plot(_solve,-PC_gaussian[:,n])
plt.axvline(0_pca,color='black',linestyle='--')
plt.legend(['cosine (%d components)'%Nmax_cos,
            'square (%d components)'%Nmax_square,
            'gaussian (%d components)'%Nmax_gaussian,
           'basis start'])
plt.savefig('plots/CV_only/' + 'PC-'+str(n+1)+'_all_lin.pdf',format='pdf')
plt.show()
```
