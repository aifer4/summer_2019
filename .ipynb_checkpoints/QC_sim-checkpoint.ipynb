```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
```

```python
"""Create a register with M qubits.""" 
M = 4
# Number of basis states is N=2**M.
N = 2**M
REG = np.ones(N, dtype=np.complex) * np.sqrt(1./N)
```

```python
def read(reg):
    """Measure the state of the register."""
    probs = np.cumsum(np.abs(reg)**2)
    x = np.random.rand()
    val = np.argmax(probs>=np.random.rand())
    # now collapse the state of the register:
    reg = np.zeros(N, dtype=np.complex)
    reg[val] = 1
    return reg, val

"""Function to split the register. This means writing states of the register as
a pair (R1,R2), where both R1 and R2 have M/2 qubits."""
def split_reg(reg):
    D = 2**(M/2)
    return np.reshape(reg,(D,D))

def wrap(reg, T):
#wraps and adds amplitudes if f is periodic.
    s = np.shape(reg)
    s_p = (s[0],T,s[1]/T)
    reg_wrapped = np.reshape(reg,s_p)
    return reg_wrapped
```
