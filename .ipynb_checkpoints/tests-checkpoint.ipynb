```python
from __future__ import division #(for python 2.7 compatibility)
import par
import modes
import spectra
import numpy as np
from numpy import log10, sqrt
import numba
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pickle
```

```python
A = np.linspace(par.a0, par.a_rec, 2*par.NT)
K = np.linspace(1,1000,par.NK)
```

```python
#solve using 3-fluid
Y3 = modes.solve_3fld(A, K)
Phi3 = Y3[:, 0, :].T
deltaG3 = Y3[:, 1, :].T
vG3 = Y3[:, 2, :].T
deltaC3 =Y3[:, 3, :].T
vC3 = Y3[:, 4, :].T
deltaN3 = Y3[:, 5, :].T
vN3 =Y3[:, 6, :].T

wC = np.zeros(2*par.NT)
cs2C = np.zeros((2*par.NT,par.NK))

wN = np.ones(2*par.NT)*1/3
cs2N = np.ones((2*par.NT,par.NK))*1/3

#solve using 3-fluid(gdm)
Y3g = modes.solve_3fld_gdm(A, K, wC, cs2C, wN, cs2N)
Phi3g = Y3g[:, 0, :].T
deltaG3g = Y3g[:, 1, :].T
vG3g = Y3g[:, 2, :].T
deltaC3g =Y3g[:, 3, :].T
vC3g = Y3g[:, 4, :].T
deltaN3g = Y3g[:, 5, :].T
vN3g =Y3g[:, 6, :].T

plt.rcParams.update(par.plot_params)

plt.plot(deltaC3.T[:,::20])
plt.gca().set_prop_cycle(None)
plt.plot(deltaC3g.T[:,::20], linestyle=':',linewidth=2)
plt.show()

plt.plot(deltaN3.T[:,::20])
plt.gca().set_prop_cycle(None)
plt.plot(deltaN3g.T[:,::20], linestyle=':',linewidth=2)
plt.show()



```

```python
"""First solve using 3-fluid to get GDM parameters"""
Y3 = modes.solve_3fld(A, K)
Phi3 = Y3[:, 0, :].T
deltaG3 = Y3[:, 1, :].T
vG3 = Y3[:, 2, :].T
deltaC3 =Y3[:, 3, :].T
vC3 = Y3[:, 4, :].T
deltaN3 = Y3[:, 5, :].T
vN3 =Y3[:, 6, :].T

OmegaB = par.OmegaB0 * A[::2]**-3
OmegaC = par.OmegaC0 * A[::2]**-3
OmegaN = par.OmegaN0 * A[::2]**-4
OmegaG = par.OmegaG0 * A[::2]**-4
OmegaD = OmegaN + OmegaC

# calculate the fiducial GDM functions
wD_F = OmegaN/(3*OmegaD)
deltaD = (OmegaC*deltaC3 + OmegaN*deltaN3)/OmegaD
vD = (OmegaC*vC3 + 4/3*OmegaN*vN3)/((1+wD_F)*OmegaD)
cs2D_F = (deltaN3*OmegaN*1/3)/(deltaD*OmegaD)

plt.rcParams.update(par.plot_params)
# Now the GDM functions can be fed back into 2-fluid:
Y2 = modes.solve_2fld(A[::2], K, wD_F, cs2D_F, deltaD[:,0], vD[:,0])

# eyeball numericall error
plt.plot(A[::2], deltaD.T[:,::20])
plt.gca().set_prop_cycle(None)
plt.plot(A[::4],Y2[:,3,::20], linestyle=':',linewidth=2)
plt.show()

plt.plot(A[::2], vD.T[:,::20])
plt.gca().set_prop_cycle(None)
plt.plot(A[::4],Y2[:,4,::20], linestyle=':',linewidth=2)
plt.show()
```

```python
delta_cs2 = np.zeros((2*par.NT,par.NK))
delta_cs2[1000:1100,:]= 0.1

Cl0 = spectra.get_Cl_3gdm(A, K, wC, cs2C, wN, cs2N)
Cl1 = spectra.get_Cl_3gdm(A, K, wC, cs2C + delta_cs2, wN, cs2N)
Cl2 = spectra.get_Cl_3gdm(A, K, wC, cs2C, wN, cs2N + delta_cs2)
plt.plot(spectra.l_list,(Cl1 - Cl0)**2)
plt.plot(spectra.l_list,(Cl2 - Cl0)**2)
plt.legend([r'perturbed $c_{sC}^2$','perturbed $c_{s\\nu}^2$'])
plt.semilogx()
```

```python
C_res = np.zeros(100)
N_res = np.zeros(100)
a_vals = np.zeros(100)

for i in range(100):
    print(i)
    a_vals[i] =A[i*2*par.NT//100]
    n_start = i*2*par.NT//100
    n_end = (i+1)*2*par.NT//100
    delta_cs2 = np.zeros((2*par.NT,par.NK))
    delta_cs2[n_start:n_end,:]= 0.1
    Cl1 = spectra.get_Cl_3gdm(A, K, wC, cs2C + delta_cs2, wN, cs2N)
    Cl2 = spectra.get_Cl_3gdm(A, K, wC, cs2C, wN, cs2N + delta_cs2)
    C_res[i]=np.sum((Cl1-Cl0)**2)
    N_res[i]=np.sum((Cl2-Cl0)**2)
    
    
```

```python
plt.semilogy()
plt.plot(a_vals,C_res)
plt.plot(a_vals,N_res)
a_CNeq = A[::2][np.argmin(abs(OmegaC - OmegaN))]

#plt.plot(A[::2],OmegaC)
#plt.plot(A[::2],OmegaN)
plt.axvline(a_CNeq,linestyle=':',color='black')
plt.axvline(par.a_eq, linestyle=':')
plt.ylabel('$\sum (C_{\ell F} - C_\ell)^2$')
plt.legend([r'perturbed $c_{sC}^2$','perturbed $c_{s\\nu}^2$','$\Omega_M = \Omega_R$','$\Omega_c = \Omega_\\nu$'])
plt.xlabel('a')
```
