# Coordinate Systems
A coordinate system is method which can be used to associate with a point in space a number or set of numbers which is unique to that point.

# Lines and Planes
We will use the following terminology:
* A coordinate line for some coordinate $\alpha$ is the set of points for which all coordinates but $\alpha$ are held to some fixed values.

* A plane (in any number of dimensions) is the set of points for which some coordinate takes a fixed value.

In two dimensions, x-coordinate lines coincide with y-coordinate planes and vice versa, but this is not true in general.

```python
import numpy as np
import matplotlib.pyplot as plt

#plot some x coordinate lines
xlines = np.array([(x,y) for x in range(100) for y in range(0,101,10)])
plt.scatter(xlines[:,0],xlines[:,1])

#plot some y coordinate lines
ylines = np.array([(x,y) for y in range(100) for x in range(0,101,10)])
plt.scatter(ylines[:,0],ylines[:,1])
plt.show()
```

```python
def T(x, y):
    theta = np.arctan2(y,x)
    r = np.sqrt(x**2 + y**2)
    return (r,theta)
xlines_p = np.array([T(x,y) for (x,y) in xlines])
ylines_p = np.array([T(x,y) for (x,y) in ylines])
plt.scatter(xlines_p[:,0],xlines_p[:,1])
plt.scatter(ylines_p[:,0],ylines_p[:,1])
```
